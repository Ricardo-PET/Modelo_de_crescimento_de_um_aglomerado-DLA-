**Considerações iniciais**

Uma "cluster" pode ser definida como sendo um aglomerado de partículas, e a "seed" como uma semente inicial que ajuda a construir este aglomerado. O modelo DLA é constintuido de uma série de condições usadas para gerar uma "cluster", estas condições são as seguintes. Considere um espaço de duas dimensões com coordenadas ($x$,$y$), em que $x$ e $y$ são números inteiros. A "cluster" é formada por uma ou mais "seeds", que vão estar em pontos fixos do espaço. As "seeds" servem para agregar partículas que entram no seu perímetro (as oito posições ao seu redor). Uma particula é um ponto no espaço capaz de se movimentar de forma aleatória, uma unidade para cima, baixo, esquerda ou direita. Quando uma partícula chega no perímetro de uma "seed", ela para de se movimentar, como se tivesse grudada nela. A partir daí essa particula se transforma em uma nova "seed", ou seja, se tranforma em uma parte da "cluster", e vai agregar qualquer partícula que entre no seu perímetro.

As partículas procuram simular o movimento encontrado em sistemas que há a difusão de moléculas, como dentro de uma solução gasosa ou líquida. A partícula não simula perfeitamente situações em que, por exemplo, exite uma concentração distoante de partículas em uma determinada região do espaço, ou se houver uma correnteza que dê velocidade as partículas em uma direção específica, entretanto serve bem para os propósitos do projeto.

Este modelo de "cluster" é inspirado em modelos de formação de estruras como o floco de neve, ou as partículas de fuligem. Por exemplo, o floco de neve é formado por um grão de poeira inicial (a "seed"), e as moléculas de água ao seu redor (as partículas) vão se agregando ao grão e criando a estrutura cristalina característica do floco de neve.

**O código**

Como vamos resolver três problemas, a "Cluster A", "Cluster B" e "Cluster C", vamos primeiro criar funções que possam ser reutilizadas nos três casos, simplificando e diminuindo o tamanho do código. Por esse motivo, nestas funções vão haver estruturas que só serão criadas logo antes delas serem chamadas, por exemplo a matriz A, ou as listas Vx e Vy.

**Importando as bibliotecas**

```python
from random import randint
from numpy import arange, zeros
from pylab import show,plot,xlabel,ylabel,title
```

**Criando a função solid**

A função solid é uma função de solidificação, ela demarca qual o perímetro de uma "seed". Para isso, vai existir uma matriz A que representa o espaço ($x$,$y$), inicialmente todas as entradas da matriz A vão estar zeradas, quando uma "seed" é criada os pontos no perímetro da posição que ela ocupa ($x_i$,$y_i$), vão receber o valor 1 na matriz A. Isso vai ajudar a saber se uma partícula está ou não no perímetro de uma "seed".

```python
def solid(xi,yi):     #Função de solidificação
    for k in range(xi-1,xi+2):
        for l in range(yi-1,yi+2):
            A[k][l] = 1
```

**Criando a função cluster**

A cluster é a principal função do código, ela vai criar as N partículas em posições aleatórias, calcular qual o movimento randômico que cada uma vai realizar, analisar se estas partículas estão no perímetro de uma "seed", e quando isso ocorrer as tranformar em "seeds". As listas "Vx" e "Vy" servem para salvar as posições em $x$ e $y$ de cada partícula no tempo atual, já as variáveis "cont" e "T" foram implementadas para parar o programa quando a "cluster" estiver completa (quando não houver nenhuma partículas se movimentando) ou quando o programa atingir um limite de tempo extremo. Para o código não ficar com um custo computacional muito alto, a checagem para saber se todas as partículas foram transformadas em "seeds" ocorrem somente a cada 500 passos do programa. A proxima posição de cada partícula é decidida pela função randint(), que dá um valor de 0 a 3 para o parâmetro "n" e a partir do resultado a partícula se move. Para não haver problemas quando uma partícula se aproxima das bordas da área simulada, foram criadas condições que impedem elas de ultrapassarem esses limites. Além de tudo isso, essa função faz um plot de como o sistema está no início da simulação.


```python
def cluster():     #Função geradora da cluster       
    for i in range(N):     #Função geradora das partículas
        Vx.append(randint(1,L-1))
        Vy.append(randint(1,L-1))
    xlabel("x")
    ylabel("y")
    title("Sitema inicial")
    plot(Vx,Vy,'b.')
    show() 
    cont = 0     #Tempo do programa
    T = 0     #Número de partículas tranformadas em seeds
    
    while (cont< 10**6 and T != N):
        T = 0
        for j in range(N):
            if (A[Vx[j]][Vy[j]] == 0):
                n = randint(0,3)
                if n == 0 and Vx[j]<L-1:    #Andar para a direita
                     Vx[j]+=1
                elif n == 1 and Vx[j]>1:    #Andar para a esquerda
                     Vx[j]-=1
                elif n == 2 and Vy[j]<L-1:    #Andar para cima
                     Vy[j]+=1
                elif n == 3 and Vy[j]>1:    #Andar para baixo
                     Vy[j]-=1
            else:
                solid(Vx[j],Vy[j])
        cont+=1
        if cont%500 == 0:
            for i in range(N):
                T+=A[Vx[i]][Vy[i]]
```

**Cluster A**

Para a "Cluster A" vamos considerar apenas uma "seed" ($x_i$,$y_i$) no centro de uma região quadrada de lado "L", e liberar um número "N" de partículas em posições aleatórias dentro da região. Com o passar do tempo estas partículas vão se agregar na "seed" e formar uma "cluster" mais complexa. No final vamos plotar o resultado com a "Cluster A" formada.


```python
L = 100    #Número par              
A = zeros([L+1,L+1],int)
N = 800     #Números de partículas 
Vx = []
Vy = []

xi,yi = int(L/2),int(L/2)
solid(xi,yi)
plot(xi,yi,'r.')

cluster()

xlabel("x")
ylabel("y")
title("Cluster A")
plot(Vx,Vy,'b.')
show() 
```

**Cluster B**

Para a "Cluster B" vamos criar duas paredes formadas por "seeds", ambas as retas vão ser paralelas ao eixo y e vão estar dentro da caixa de lado "L", uma delas na posição x1 = 20 e outra na x2 = 80. Como no caso anterior, também vamos liberar um número "N" de partículas com movimentos aleatórios dentro deste espaço. No final vamos plotar o resultado com a "Cluster B" formada.

Finalmente, para a "Cluster C" vamos considerar apenas uma "seed" no ponto (1,1) da região quadrada de lado "L", e liberar um número "N" de partículas em posições aleatórias dentro da região. A cluster formada neste caso vai ter uma configuração diferenciada pois a "seed" inicial só poderá agregar parículas vindas de cima ou da sua direita. No final vamos plotar o resultado com a "Cluster C" formada.

```python
L = 100    #Número par              
A = zeros([L+1,L+1],int)
N = 800     #Números de partículas 
Vx = []
Vy = []

xi,yi = 1,1
solid(xi,yi)
plot(xi,yi,'r.')

cluster()

xlabel("x")
ylabel("y")
title("Cluster C")
plot(Vx,Vy,'b.')
show() 
```
